```python

def Rosetta_Threshold(Interface_score,Label):
    
    max_score = max(Interface_score)
    min_score =min(Interface_score)    
    cutoff_mattews_pair_list =[]
    best_mattews = 0.0000000000000000000000000000000000000001
    best_acc = 0.0000000000001
    best_cutoff = 0
    mattews_score_list = []
    pred_label=[0 for i in range(len(Label))]
 
    
    for cutoff in linspace(-50,3,500):
        temp_pred_label=[]
        true_positive =0
        false_negative =0
        true_negative = 0
        false_positive = 0
        
        for j in range(len(Interface_score)):
            
            if (Interface_score[j] <= cutoff):
            #if (Interface_score[j] >= cutoff):
                if (Label[j] == 1):
                    true_positive += 1.
                else:
                    false_positive +=1.
                temp_pred_label.append(1)
            else:
                if (Label[j] == 1):
                    false_negative += 1.
                else:
                    true_negative +=1.
                temp_pred_label.append(0)
        #f1_score_1 = 2.*true_positive/(2.*true_positive + false_positive + false_negative)
        #f1_score_0 = 2.*true_negative/(2.*true_negative + false_negative + false_positive)
        accuracy = (true_positive + true_negative)/(true_positive + true_negative + false_positive + false_negative)
        mattews = (true_positive*true_negative + false_positive*false_negative)
        mattews = mattews * 1./np.sqrt((true_positive + false_positive)*(true_positive + false_negative)*(true_negative + false_positive)*(true_negative + false_negative))
        #f1_score = (f1_score_0 + f1_score_1)/2.
        mattews_score_list.append(mattews)
        if (mattews > best_mattews):
            best_mattews = mattews
            best_acc = accuracy
            best_cutoff =cutoff
            best_tp = true_positive
            best_tn = true_negative
            best_fn = false_negative
            best_fp = false_positive
            if len(pred_label)==len(temp_pred_label):
                pred_label=temp_pred_label
            else:
                print ("Some predictions are missing")
                print (len(pred_label),len(temp_pred_label),temp_pred_label)
                    
        #print cutoff,mattews
        #print "true_positive   true negative   false_negative    false_positive"
        #print true_positive,true_negative,false_negative,false_positive
        #print "//////////////////////////////////////////////////////////////////////////"
        #cutoff_f1 = [cutoff,f1_score]
        #cutoff_f1_pair_list.append(cutoff_f1)
    #plt.plot(linspace(min_score,100,200),f1_score_list)
    #plt.plot(linspace(-10,2,200),mattews_score_list)
    #plt.title("Rosetta Threshold", fontsize = 14)
    #xlab = 'Rosetta Inteface Energy'
    #ylab = 'Mathews-Score'
    #plt.xlabel(xlab, fontsize=12)
    #plt.ylabel(ylab, fontsize=12)
    #plt.show()
    #print "//////////////////////////////////////////////////////////////////////////"
    #print "//////////////////////////////////////////////////////////////////////////"
    #print "best_cutoff,best_mattews,best_tp,best_tn,best_fn,best_fp"
    #print best_cutoff,best_mattews,best_tp,best_tn,best_fn,best_fp
    result = [best_cutoff,best_mattews,best_tp,best_tn,best_fn,best_fp,best_acc,pred_label]
    return result
```

```python
def prediction_from_cutoff(cutoff,Interface_score,Label):
        pred_label=[]
        true_positive =0
        false_negative =0
        true_negative = 0
        false_positive = 0
        
        for j in range(len(Interface_score)):
            
            if (Interface_score[j] <= cutoff):
            #if (Interface_score[j] >= cutoff):
                if (Label[j] == 1):
                    true_positive += 1.
                else:
                    false_positive +=1.
                pred_label.append(1)
            else:
                if (Label[j] == 1):
                    false_negative += 1.
                else:
                    true_negative +=1.
                pred_label.append(0)
        #f1_score_1 = 2.*true_positive/(2.*true_positive + false_positive + false_negative)
        #f1_score_0 = 2.*true_negative/(2.*true_negative + false_negative + false_positive)
        accuracy = (true_positive + true_negative)/(true_positive + true_negative + false_positive + false_negative)
        mattews = (true_positive*true_negative + false_positive*false_negative)
        mattews = mattews * 1./np.sqrt((true_positive + false_positive)*(true_positive + false_negative)*(true_negative + false_positive)*(true_negative + false_negative))
        #f1_score = (f1_score_0 + f1_score_1)/2.
        
        result = [mattews,true_positive,true_negative,false_negative,false_positive,accuracy,pred_label]
        return result
```

```python
def stratified_cv(X, y, clf_class, random_state=None,shuffle=False, n_folds=10, **kwargs):
    # Construct a kfolds object
    #stratified_k_fold = cross_validation.StratifiedShuffleSplit(y, 10, test_size=0.2,)
    stratified_k_fold = cross_validation.StratifiedKFold(y, n_folds=n_folds,shuffle=shuffle,random_state=10) 
    y_pred = y.copy()
    for train_index, test_index in stratified_k_fold:
        X_train, X_test = X[train_index], X[test_index]
        y_train = y[train_index]
        clf = clf_class(**kwargs)
        clf.fit(X_train,y_train)
        y_pred[test_index] = clf.predict(X_test)
    return y_pred
```

```python
def stratified_cv_proba(X, y, clf_class, random_state=None,shuffle=False, n_folds=10, **kwargs):
    # Construct a kfolds object
    #stratified_k_fold = cross_validation.StratifiedShuffleSplit(y, 10, test_size=0.2,)
    stratified_k_fold = cross_validation.StratifiedKFold(y, n_folds=n_folds,shuffle=shuffle,random_state=10) 
    y_pred = y.copy().astype(float)
    for train_index, test_index in stratified_k_fold:
        X_train, X_test = X[train_index], X[test_index]
        y_train = y[train_index]
        clf = clf_class(**kwargs)
        clf.fit(X_train,y_train)
        y_pred[test_index] = clf.predict_proba(X_test)[:,1]
    return y_pred
```

```python
def stratified_cv_rosetta(X, y, random_state=None,shuffle=False, n_folds=10, **kwargs):
    
    stratified_k_fold = cross_validation.StratifiedKFold(y, n_folds=10,shuffle=False,random_state=10)
    
    y_pred = y.copy()
    cutoff_list =[]
    for train_index, test_index in stratified_k_fold:
        X_train = X[train_index]
        y_train = y[train_index] 
        X_test = X[test_index]
        y_test = y[test_index]
        
        cutoff = Rosetta_Threshold(X_train,y_train)[0]
        cutoff_list.append(cutoff)
        y_pred[test_index] = prediction_from_cutoff(cutoff,X_test ,y_test)[6]
    return y_pred, np.mean(cutoff_list)
```

```python
#Normalization of the dictionary
def normalize_confusion_dictionary(dictionary):
    new_dict = {}
    for ii, values in dictionary.items():
        new_mat = {}
        matrix = values['matrix']
        title = values['title']
        #print (matrix.astype(float)/8000.0)
    
    
        mat_norm = np.matrix([[0., 0.],[0., 0.]])
        for i in range(2):
            for j in range(2):
                mat_norm[i,j]=matrix[i,j].astype(float)/(matrix[:,j].sum()).astype(float)
        new_mat['matrix'] = mat_norm 
        new_mat['title'] = title
        new_dict[ii] = new_mat
    return new_dict
    
```

```python
def AUC_Threshold(Interface_score,Label):
    max_score = max(Interface_score)
    min_score =min(Interface_score)    

    best_mattews = 0.0000000000000000000000000000000000000001
    #best_acc = 0.0000000000001
    best_cutoff = 0

    pred_label=[0 for i in range(len(Label))]
    
    for cutoff in linspace(-50,3,500):
        temp_pred_label=[]
        true_positive =0
        false_negative =0
        true_negative = 0
        false_positive = 0
    
        for j in range(len(Interface_score)):

            #if (Interface_score[j] >= cutoff):
            if (Interface_score[j] <= cutoff):
            #if (Interface_score[j] >= cutoff):
                if (Label[j] == 1):
                    true_positive += 1.
                else:
                    false_positive +=1.
                temp_pred_label.append(1)
            else:
                if (Label[j] == 1):
                    false_negative += 1.
                else:
                    true_negative +=1.
                temp_pred_label.append(0)
    
        mattews = (true_positive*true_negative + false_positive*false_negative)
        mattews = mattews * 1./np.sqrt((true_positive + false_positive)*(true_positive + false_negative)*(true_negative + false_positive)*(true_negative + false_positive))
        #f1_score = (f1_score_0 + f1_score_1)/2.
        
        if (mattews > best_mattews):
            best_mattews = mattews
            #best_acc = accuracy
            best_cutoff =cutoff
            if len(pred_label)==len(temp_pred_label):
                pred_label=temp_pred_label
            else:
                print "Some predictions are missing"
                print len(pred_label),len(temp_pred_label)
           
    return Label, pred_label
```

```python
def confusion_mat_score(score, cutoff, df, label):
    if score != 'vina_affinity':
        pred = (df[score]>=cutoff).astype(int)
    else:
        pred = (df[score]<=cutoff).astype(int)
    return metrics.confusion_matrix(df[label].values,pred)
    
```

```python
def matthews_corr_coeff(score, cutoff, df, label):
    if score != 'vina_affinity':
        pred = (df[score]>=cutoff).astype(int)
    else:
        pred = (df[score]<=cutoff).astype(int)
    return metrics.matthews_corrcoef(df[label].values,pred)
```

```python
def class_report(score, cutoff, df, label):
    if score != 'vina_affinity':
        pred = (df[score]>=cutoff).astype(int)
    else:
        pred = (df[score]<=cutoff).astype(int)
    return metrics.classification_report(df[label].values,pred)
```

```python
def precision_recall_mcc_from_cutoff(cutoff,Interface_score,Label):
        pred_label=[]
        true_positive =0
        false_negative =0
        true_negative = 0
        false_positive = 0
        
        for j in range(len(Interface_score)):
            
            if (Interface_score[j] <= cutoff):
            #if (Interface_score[j] >= cutoff):
                if (Label[j] == 1):
                    true_positive += 1.
                else:
                    false_positive +=1.
                pred_label.append(1)
            else:
                if (Label[j] == 1):
                    false_negative += 1.
                else:
                    true_negative +=1.
                pred_label.append(0)
        #f1_score_1 = 2.*true_positive/(2.*true_positive + false_positive + false_negative)
        #f1_score_0 = 2.*true_negative/(2.*true_negative + false_negative + false_positive)
        #accuracy = (true_positive + true_negative)/(true_positive + true_negative + false_positive + false_negative)
        #mattews = (true_positive*true_negative + false_positive*false_negative)
        #mattews = mattews * 1./np.sqrt((true_positive + false_positive)*(true_positive + false_negative)*(true_negative + false_positive)*(true_negative + false_negative))
        precision = metrics.precision_score(Label, pred_label)
        recall = metrics.recall_score(Label, pred_label)
        matthews = metrics.matthews_corrcoef(Label,pred_label)
        #f1_score = (f1_score_0 + f1_score_1)/2.
        
        result = [matthews, precision,recall]
        return result
```

```python

def ML_Threshold(ML_score,Label):
    
    max_score = max(ML_score)
    #print (max_score)
    min_score =min(ML_score)  
    #print (min_score)
    cutoff_mattews_pair_list =[]
    best_mattews = 0.0000000000000000000000000000000000000001
    best_acc = 0.0000000000001
    best_cutoff = 0
    mattews_score_list = []
    pred_label=[0 for i in range(len(Label))]
 
    
    for cutoff in linspace(min_score,max_score,20):
        #print(cutoff)
        temp_pred_label=[]
        true_positive =0
        false_negative =0
        true_negative = 0
        false_positive = 0
        
        for j in range(len(ML_score)):
            
            if (ML_score[j] >= cutoff):
                
            #if (Interface_score[j] >= cutoff):
                if (Label[j] == 1):
                    true_positive += 1.
                    #print(ML_score[j])
                    #print ('cufoff', cutoff)
                else:
                    false_positive +=1.
                temp_pred_label.append(1)
            else:
                if (Label[j] == 1):
                    false_negative += 1.
                else:
                    true_negative +=1.
                temp_pred_label.append(0)
        #f1_score_1 = 2.*true_positive/(2.*true_positive + false_positive + false_negative)
        #f1_score_0 = 2.*true_negative/(2.*true_negative + false_negative + false_positive)
        accuracy = (true_positive + true_negative)/(true_positive + true_negative + false_positive + false_negative)
        mattews = (true_positive*true_negative + false_positive*false_negative)
        mattews = mattews * 1./(np.sqrt((true_positive + false_positive)*(true_positive + false_negative)*(true_negative + false_positive)*(true_negative + false_negative))+1e-10)
        #print(mattews)
        #f1_score = (f1_score_0 + f1_score_1)/2.
        mattews_score_list.append(mattews)
        if (mattews > best_mattews):
            best_mattews = mattews
            best_acc = accuracy
            best_cutoff =cutoff
            best_tp = true_positive
            best_tn = true_negative
            best_fn = false_negative
            best_fp = false_positive
            if len(pred_label)==len(temp_pred_label):
                pred_label=temp_pred_label
            else:
                print ("Some predictions are missing")
                print (len(pred_label),len(temp_pred_label),temp_pred_label)
                    
        #print cutoff,mattews
        #print "true_positive   true negative   false_negative    false_positive"
        #print true_positive,true_negative,false_negative,false_positive
        #print "//////////////////////////////////////////////////////////////////////////"
        #cutoff_f1 = [cutoff,f1_score]
        #cutoff_f1_pair_list.append(cutoff_f1)
    #plt.plot(linspace(min_score,100,200),f1_score_list)
    #plt.plot(linspace(-10,2,200),mattews_score_list)
    #plt.title("Rosetta Threshold", fontsize = 14)
    #xlab = 'Rosetta Inteface Energy'
    #ylab = 'Mathews-Score'
    #plt.xlabel(xlab, fontsize=12)
    #plt.ylabel(ylab, fontsize=12)
    #plt.show()
    #print "//////////////////////////////////////////////////////////////////////////"
    #print "//////////////////////////////////////////////////////////////////////////"
    #print "best_cutoff,best_mattews,best_tp,best_tn,best_fn,best_fp"
    #print best_cutoff,best_mattews,best_tp,best_tn,best_fn,best_fp
    result = [best_cutoff,best_mattews,best_tp,best_tn,best_fn,best_fp,best_acc,pred_label]
    return result
```

```python
def ml_prediction_from_cutoff(cutoff,ML_score,Label):
        pred_label=[]
        true_positive =0
        false_negative =0
        true_negative = 0
        false_positive = 0
        
        for j in range(len(ML_score)):
            
            if (ML_score[j] >= cutoff):
            #if (Interface_score[j] >= cutoff):
                if (Label[j] == 1):
                    true_positive += 1.
                else:
                    false_positive +=1.
                pred_label.append(1)
            else:
                if (Label[j] == 1):
                    false_negative += 1.
                else:
                    true_negative +=1.
                pred_label.append(0)
        #f1_score_1 = 2.*true_positive/(2.*true_positive + false_positive + false_negative)
        #f1_score_0 = 2.*true_negative/(2.*true_negative + false_negative + false_positive)
        accuracy = (true_positive + true_negative)/(true_positive + true_negative + false_positive + false_negative)
        mattews = (true_positive*true_negative + false_positive*false_negative)
        mattews = mattews * 1./np.sqrt((true_positive + false_positive)*(true_positive + false_negative)*(true_negative + false_positive)*(true_negative + false_negative))
        #f1_score = (f1_score_0 + f1_score_1)/2.
        
        result = [mattews,true_positive,true_negative,false_negative,false_positive,accuracy,pred_label]
        return result
```

```python
def stratified_cv_ML(X, y, random_state=None,shuffle=False, n_folds=10, **kwargs):
    
    stratified_k_fold = cross_validation.StratifiedKFold(y, n_folds=10,shuffle=False,random_state=10)
    
    y_pred = y.copy()
    cutoff_list =[]
    for train_index, test_index in stratified_k_fold:
        X_train = X[train_index]
        y_train = y[train_index] 
        X_test = X[test_index]
        y_test = y[test_index]
        
        cutoff = ML_Threshold(X_train,y_train)[0]
        cutoff_list.append(cutoff)
        y_pred[test_index] = ml_prediction_from_cutoff(cutoff,X_test ,y_test)[6]
    return y_pred, np.mean(cutoff_list)
```
